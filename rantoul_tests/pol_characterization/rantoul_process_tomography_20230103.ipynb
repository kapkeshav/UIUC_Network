```python
def extract_data(lh,ld,lt,d):
    vh = d[4]/d[3]
    pd = d[5]/(d[6]+d[5])
    pa = d[6]/(d[6]+d[5])
    lt.append(datetime.datetime.fromtimestamp(d[2]))
    lh.append(vh)
    ld.append((pd-pa)/(2*np.sin(np.arctan(vh))))
```
