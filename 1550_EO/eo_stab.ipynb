```python
samples = np.array([d[0] for d in data])
BS1_0 = np.array([d[1] for d in data])
BS0_0 = np.array([d[2] for d in data])
BS1_1 = np.array([d[3] for d in data])
```

```python
plt.scatter(samples/60/60,(BS0_0+BS1_0+BS1_1)*1e6,label=r"Mean= {}, $\sigma$ = {}$\mu W$".format(np.average((BS0_0+BS1_0+BS1_1)*1e6),np.std((BS0_0+BS1_0+BS1_1)*1e6)))
total_power = (BS0_0+BS1_0+BS1_1)*1e6
average_power = [np.average(total_power[i*562:(i+1)*562]) for i in range(32)]
plt.plot(samples[::562]/60/60,average_power,color='orange',label="averaged power per 562 seconds")
plt.xlabel("Time (h)")
plt.ylabel(r"Power ($\mu W$)")
plt.title("EO-Converter Total Power")
plt.legend()
plt.savefig("EO_power_20221027.pdf")
```

```python
plt.plot(samples/60/60,BS1_0/(BS1_0+BS1_1))
plt.plot(samples/60/60,BS1_1/(BS1_0+BS1_1))
plt.ylabel("Portion of Power")
plt.xlabel("Time (h)")
plt.title("Paddle Stability for EO-converter")
plt.savefig("EO_pol_20221027.pdf")
```
